🧹 Step 4: Data Cleaning
Handle Missing Values

```python
# Fill missing values
df['director'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)
```

📈 Step 6: Advanced Analysis

```python
# Movies Released After 2015
movies_2015 = df[(df['release_year'] > 2015) & (df['type'] == 'Movie')]
print(movies_2015[['title', 'release_year']].head())
```

```python
# Find Longest Movie 
movie_data = df[df['type'] == 'Movie']

movie_data['duration'] = movie_data['duration'].str.replace(' min','')
movie_data['duration'] = movie_data['duration'].astype(float)

print(movie_data[['title','duration']].sort_values(by='duration', ascending=False).head())
```

```python
# Content Added by Year
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(12,6))
sns.countplot(x='year_added', data=df,color='maroon')
plt.xticks(rotation=42)
plt.title('Content Added by Year')
plt.show()
```

📊 Step 7: Correlation Heatmap

```python
#📊 Step 7: Correlation Heatmap
numeric_data = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(8,5))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
